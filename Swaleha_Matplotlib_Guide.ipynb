# Matplotlib in Python
- Matplotlib is a plotting library for the Python programming language
- Pyplot functions help creating a figure, plotting area and decorates the plot with labels

```
import matplotlib.pyplot as plt
```

```
plt.plot([1, 2, 3, 4]) # Data Points
plt.ylabel('Numbers') # labelling y-axis
plt.xlabel('Indices') # labelling x-axis
plt.title("My Graph") # Title of the plot
plt.show()
```

```
# Plot with Grids
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Graph')
plt.grid()
plt.show()
```

```
# Plot with dot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.grid()
plt.show()
```

```
import numpy as np
t = np .arange(1., 5., 0.2)

plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**3, 'rs', label='^3')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

### Line Properties

```
# Changing line width
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth='4.0')
plt.grid()
plt.show()
```

```
# setp function in matplotlib

x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [3, 6, 9, 12]

lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines[0], color='r', linewidth='2.0') # Change specs of each line
plt.setp(lines[1], color='g', linewidth='5.0')
plt.grid()
plt.show()

```

### Handling multiple figure and axes

```
t1 = np.arange(1.0, 5.0, 0.1)
t2 = np.arange(1.0, 5.0, 0.3)

plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1, t1**2, 'ro')

plt.subplot(212)
plt.plot(t2, np.sin(t2), 'b--')
plt.show()
```
